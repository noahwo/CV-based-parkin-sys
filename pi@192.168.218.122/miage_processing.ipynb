```python
import matplotlib
import cv2
import re
from matplotlib import pyplot as plt
import numpy as np
```

```python
img = cv2.imread("img.jpg")

scale_percent = 40  # percent of original size
width = int(img.shape[1] * scale_percent / 100)
height = int(img.shape[0] * scale_percent / 100)
dim = (width, height)

# resize image
resized = cv2.resize(img, dim, interpolation=cv2.INTER_AREA)
```

```python
img = resized
plt.imshow(img)
cv2.imwrite("orig_img.jpg",img)

imgGray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(imgGray)
cv2.imwrite("gray_img.jpg",imgGray)
```

```python
imgBlur = cv2.GaussianBlur(imgGray, (5, 5), 1)

plt.imshow(imgBlur)
cv2.imwrite("blur_img.jpg",imgBlur)
```

```python
imgThreshold = cv2.adaptiveThreshold(
    imgBlur, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 25, 16
)
plt.imshow(imgThreshold)

cv2.imwrite("thres_img.jpg",imgThreshold)
```

```python
imgMedian = cv2.medianBlur(imgThreshold, 7)

plt.imshow(imgMedian)
cv2.imwrite("median_img.jpg",imgMedian)


```

```python
# kernel = np.ones((3, 3), np.uint8)
# kernel = np.ones((5, 5), np.uint8)
kernel = np.ones((7, 7), np.uint8)
imgDilate = cv2.dilate(imgMedian, kernel, iterations=1)

plt.imshow(imgDilate)
cv2.imwrite("dilate_img.jpg",imgDilate)


```
